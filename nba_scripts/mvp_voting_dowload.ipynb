```python
# download the mvp voting dataset from kaggle
import kagglehub

path = kagglehub.dataset_download("parthdande/nba-mvp-voting-dataset-2000-2021")
print("Path to dataset files:", path)
```

```python
# move the csv files of the mvp voting into the current data directory
import shutil
import os

destination_folder = "./data"
if not os.path.exists(destination_folder):
    os.makedirs(destination_folder)

for filename in os.listdir(path):
    source_file = os.path.join(path, filename)
    destination_file = os.path.join(destination_folder, filename)
    
    shutil.copy(source_file, destination_file)
    print(f"Successfully moved: {filename} to {destination_folder}")
```

```python
import os

def remove_spaces_from_filenames(directory):
    # Check if the directory exists
    if not os.path.exists(directory):
        print(f"‚ùå Error: The directory '{directory}' does not exist.")
        return

    print(f"üìÇ Scanning directory: {directory}...")
    
    # Counter for renamed files
    count = 0

    # Iterate through all files in the specified directory
    for filename in os.listdir(directory):
        # Construct the full file path
        old_path = os.path.join(directory, filename)

        # Skip if it's a directory, we only want to rename files
        if os.path.isdir(old_path):
            continue

        # Check if there is a space in the filename
        if ' ' in filename:
            # Create new filename by removing all spaces
            new_filename = filename.replace(' ', '')
            new_path = os.path.join(directory, new_filename)

            # Safety check: avoid overwriting existing files
            if os.path.exists(new_path):
                print(f"‚ö†Ô∏è Warning: Cannot rename '{filename}' because '{new_filename}' already exists.")
                continue

            try:
                os.rename(old_path, new_path)
                print(f"‚úÖ Renamed: '{filename}' ‚û°Ô∏è '{new_filename}'")
                count += 1
            except Exception as e:
                print(f"‚ùå Failed to rename '{filename}': {e}")

    print(f"\n‚ú® Task complete. Total files renamed: {count}")


target_folder = 'data'
remove_spaces_from_filenames(target_folder)
```
